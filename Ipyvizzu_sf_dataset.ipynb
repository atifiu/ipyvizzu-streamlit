```python
import pandas as pd
from ipyvizzu import Chart, Data, Config, Style
import requests

#Add Titanic data set to a data frame with a Count value added to it
data_frame = pd.read_csv(
    "https://data.sfgov.org/resource/rkru-6vcg.csv"
)
data = Data()
data.add_data_frame(data_frame)

data_frame.head(4)
```

```python
top10_airlines = (data_frame.groupby('operating_airline').agg(total_pass=('passenger_count', 'sum')).sort_values(by = 'total_pass', ascending = False)
 .nlargest(10, 'total_pass').reset_index()['operating_airline'].tolist())
```

```python
chart = Chart(width="80%")

chart.animate(data)
chart.animate(
    Config(
        {
            "x": "geo_region",
            "y": ["passenger_count"],
            "color": "geo_region",
            "geometry": "rectangle",
            "align": "none",
            "title": "Passenger count by Region",
            "sort":"byValue",
            "reverse": False,
        }
    ),
    Style(
        {
            "plot": {
                "xAxis": {"label": {"fontSize": 12, "angle": 2.0}},
                "marker": {
                     "colorPalette": "#d9ed92ff #b5e48cff #99d98cff #76c893ff #52b69aff"
                    + " #34a0a4ff #168aadff #1a759fff #1e6091ff #184e77ff"
                },"paddingBottom": 110,
            }, "title": {"paddingTop": 0}
        }
    ),delay=3,
)
#to enable tooltip
chart.feature("tooltip", True)

# horizontal column using groupedBar
chart.animate(
    Config.groupedBar(
        {
            #"align": "stretch",
            "x": "passenger_count",
            "y": ["geo_region"],
            "groupedBy": "geo_region",
            #"color": ["geo_region"],
            #"title": "Passenger count by Region(%)",
        }
    ),
    Style(
        {
            "plot": {
                "xAxis": {
                        "label": {"fontSize": 12, "angle": 2.0}, 
                         "title": {"paddingTop": 50}
                        }, "paddingLeft":120
                
        }
        }
    ),
    delay=3,
)

#Bubble Plot
chart.animate(
    Config.bubble(
        {
            "size": "passenger_count",
            "color": "geo_region",
            "title": "Passenger count distribution by Region",
        }
    ),delay=3,
)

#treemap
chart.animate(
    Config.treemap(
        {
            "size": "passenger_count",
            "color": "geo_region",
            "title": "Passenger count distribution by Region",
        }
    ),delay=3,
)
```

```python
#top 10 airlines by passenger count


#chart.animate(data)

filter_format = "(" + \
    " || ".join([f"record['operating_airline'] == '{item}'" for item in top10_airlines]) + ")"

#print(filter_format)

chart.animate(data.filter(filter_format))
chart.animate(
    Config.groupedBar(
        {
            #"align": "stretch",
            "x": "passenger_count",
            "y": ["operating_airline"],
            "groupedBy": "operating_airline",
            #"color": ["geo_region"],
            "title": "Top 10 Airlines by Passenger Count",
            "sort":"byValue",
            
        }
    ),
    Style(
        {
            "plot": {
                "xAxis": {"label": {"fontSize": 12, "angle": 2.0}},
                "marker": {
                    "colorPalette": "#d9ed92ff #b5e48cff #99d98cff #76c893ff #52b69aff"
                    + " #34a0a4ff #168aadff #1a759fff #1e6091ff #184e77ff"
                },"paddingBottom": 110,
            }, "title": {"paddingTop": 0}
        }
    ), delay = 3
)

#Bubble Plot
chart.animate(
    Config.bubble(
        {
            "size": "passenger_count",
            "color": "operating_airline",
            "title": "Passenger count distribution by Airlines",
        }
    ),delay=3,
)

#treemap
chart.animate(
    Config.treemap(
        {
            "size": "passenger_count",
            "color": "operating_airline",
            "title": "Passenger count distribution by Airlines",
        }
    ),delay=3,
)
#to enable tooltip
#chart.feature("tooltip", True)
```
